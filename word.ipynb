```python
from typing import List 
import collections

class Solution:
    def uncommonFromSentences(self, s1: str, s2: str) -> List[str]:
        wordslist = s1.split(" ") + s2.split(" ")
        words = collections.Counter(wordslist) #

        new = []

        for word in words:
            if words[word]==1:
                new.append(word)
        return new

a = Solution()
s1 = "apple apple"
s2 = "banana"

print(a.uncommonFromSentences(s1,s2))

#389 Find the difference 
#1935 Maximum number of words you can type 
```

```python
#Find the difference 

#You are given two strings s and t.
#String t is generated by random shuffling string s and then add one more letter at a random position.
#Return the letter that was added to t.

#Input: s = "abcd", t = "abcde"
#Output: "e"
#Explanation: 'e' is the letter that was added.

#Input: s = "", t = "y"
#Output: "y"
class Solution:
    def findTheDifference(self, s:str, t:str) -> str:
        for i in t:
            if s.count(i) != t.count(i):
                return i

a = Solution()
print(a.findTheDifference("abcd","abcde"))
print(a.findTheDifference("", "y"))
```

```python
#There is a malfunctioning keyboard where some letter keys do not work. All other keys on the keyboard work properly.
#Given a string text of words separated by a single space (no leading or trailing spaces) and a string brokenLetters of all distinct letter keys that are broken, 
#return the number of words in text you can fully type using this keyboard.
# Input: text = "hello world", brokenLetters = "ad"
#Output: 1
#Explanation: We cannot type "world" because the 'd' key is broken.
#Input: text = "leet code", brokenLetters = "lt"
#Output: 1
#Explanation: We cannot type "leet" because the 'l' and 't' keys are broken.
#Input: text = "leet code", brokenLetters = "e"
#Output: 0
#Explanation: We cannot type either word because the 'e' key is broken.

class Solution:
    def canBeTypedWords(self, text: str, brokenLetters: str) -> int:
        answer = 0 
        brokenkey = set(brokenLetters)
        for word in text.split(" "):
            if brokenkey.intersection(word):  #no common element 
                answer += 1
        return answer

a = Solution()
print(a.canBeTypedWords("hello world", "had"))
```

```python
class Solution:
    def canBeTypedWords(self, text: str, brokenLetters: str) -> int:
        answer = 0 
        text = text.split(" ")
        brokenLetters = list(brokenLetters)
        for word in text:
            if any(i in brokenLetters for i in word):
                continue
            else:
                answer += 1
        return answer

a = Solution()
print(a.canBeTypedWords("hello world this is something great", "o"))
```
